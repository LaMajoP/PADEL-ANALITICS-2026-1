## Importacion de librerias y dataset

```python
import pandas as pd
```

```python
df_raw = pd.read_csv('../datasets/dataset_padel.csv')
print(f'Dataset original: {df_raw.shape[0]:,} filas, {df_raw.shape[1]} columnas')
```

## Exploracion de datos

```python
pd.set_option('display.max_columns', None)
df_raw.head()
```

```python
df = df_raw[['cancha','partido','punto','frame',
            'j1_angulo_codo_d', 'j1_angulo_hombro_d', 'j1_angulo_inclinacion_raqueta_d', 'j1_angulo_raqueta_hombro_d', 'j1_angulo_muneca_cabeza_d',
            'j2_angulo_codo_d', 'j2_angulo_hombro_d', 'j2_angulo_inclinacion_raqueta_d', 'j2_angulo_raqueta_hombro_d', 'j2_angulo_muneca_cabeza_d',
            'j3_angulo_codo_d', 'j3_angulo_hombro_d', 'j3_angulo_inclinacion_raqueta_d', 'j3_angulo_raqueta_hombro_d', 'j3_angulo_muneca_cabeza_d',
            'j4_angulo_codo_d', 'j4_angulo_hombro_d', 'j4_angulo_inclinacion_raqueta_d', 'j4_angulo_raqueta_hombro_d', 'j4_angulo_muneca_cabeza_d']]

print(f'df shape: {df.shape}')
df.head()
```

```python
df_unpivot = df.melt(
    id_vars=['cancha', 'partido', 'punto', 'frame'],
    var_name='variable',
    value_name='valor'
)

# Extraer jugador y nombre de métrica
df_unpivot['jugador'] = df_unpivot['variable'].str.extract(r'^(j\d+)_')
df_unpivot['metrica'] = df_unpivot['variable'].str.replace(r'^j\d+_', '', regex=True)

# Pivot para dejar métricas como columnas
df_final = df_unpivot.pivot_table(
    index=['cancha', 'partido', 'punto', 'frame', 'jugador'],
    columns='metrica',
    values='valor'
).reset_index()

df_final.columns.name = None
df_final = df_final.set_index(['cancha','partido','punto','frame','jugador'])
df_final.head()
```

## Preparación de datos

```python
feature_cols = list(df_final.select_dtypes(include='number').columns)

df = df_final.dropna(subset=feature_cols)

print(f'df: {df.shape[0]:,} filas, {df.shape[1]} columnas')
df.head()
```

```python
feature_cols
```

## Escalado + Elbow method

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

scaler = StandardScaler()
X = scaler.fit_transform(df[feature_cols])

inertias = []
K_range = range(2, 11)

for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X)
    inertias.append(km.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(K_range, inertias, marker='o')
plt.xlabel('Número de clusters (k)')
plt.ylabel('Inercia')
plt.title('Elbow Method')
plt.xticks(K_range)
plt.tight_layout()
plt.show()
```

## KMeans — ajustar k según el codo

```python
K_FINAL = 3  # ajustar según el codo

kmeans = KMeans(n_clusters=K_FINAL, random_state=42, n_init=10)
kmeans.fit(X)

df['cluster'] = kmeans.labels_
print(df['cluster'].value_counts().sort_index())
```

## Visualización — PCA 2D

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2, random_state=42)
X_2d = pca.fit_transform(X)

plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_2d[:, 0], X_2d[:, 1], c=df['cluster'], cmap='tab10', alpha=0.4, s=5)
plt.colorbar(scatter, label='Cluster')
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}%)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}%)')
plt.title(f'KMeans k={K_FINAL} — proyección PCA 2D')
plt.tight_layout()
plt.show()
```

## Visualización UMAP — Golpes verificados vs clusters

Se aplica UMAP sobre los mismos vectores de ángulos escalados para proyectar
los ~158K puntos (jugador × frame) a 2D. Luego se marcan visualmente
los 81 frames de golpeo revisados manualmente para ver si se concentran
en regiones específicas del espacio.


```python
import umap
import time

df_golpes = pd.read_csv('../datasets/golpes_revisados.csv', sep='\t')

print(f"golpes_revisados: {len(df_golpes)} filas")
print(f"Con cluster asignado: {df_golpes['cluster'].notna().sum()}")
print(f"Sin cluster (no detectados): {df_golpes['cluster'].isna().sum()}")
print()

# Marcar en df cuáles filas son golpes verificados
golpe_keys = set(
    zip(df_golpes['cancha'], df_golpes['partido'],
        df_golpes['punto'],  df_golpes['frame'], df_golpes['jugador'])
)
df['es_golpe'] = [
    1 if (r.cancha, r.partido, r.punto, r.frame, r.jugador) in golpe_keys else 0
    for r in df.itertuples()
]

n_golpes_en_df = df['es_golpe'].sum()
print(f"Golpes verificados presentes en el dataset (con ángulos): {n_golpes_en_df}")
print(f"  Cluster 0: {df[(df['es_golpe']==1) & (df['cluster']==0)].shape[0]}")
print(f"  Cluster 1: {df[(df['es_golpe']==1) & (df['cluster']==1)].shape[0]}")
print(f"  Cluster 2: {df[(df['es_golpe']==1) & (df['cluster']==2)].shape[0]}")
```

```python
# Aplicar UMAP sobre X (ya escalado por StandardScaler)
print(f"Aplicando UMAP a {X.shape[0]:,} puntos x {X.shape[1]} features...")
t0 = time.time()

reducer = umap.UMAP(
    n_components=2,
    n_neighbors=30,
    min_dist=0.05,
    random_state=42,
    n_jobs=-1,
)
embedding = reducer.fit_transform(X)

print(f"UMAP completado en {time.time()-t0:.1f}s")

df['umap_x'] = embedding[:, 0]
df['umap_y'] = embedding[:, 1]
```

```python
CLUSTER_COLORS = {0: '#3498db', 1: '#2ecc71', 2: '#f39c12'}
CLUSTER_LABELS = {0: 'Cluster 0', 1: 'Cluster 1', 2: 'Cluster 2'}

fig, ax = plt.subplots(figsize=(12, 9))

# ── Fondo: todos los puntos, pequeños, coloreados por cluster ──
for cl in sorted(df['cluster'].unique()):
    mask = (df['cluster'] == cl) & (df['es_golpe'] == 0)
    ax.scatter(
        df.loc[mask, 'umap_x'],
        df.loc[mask, 'umap_y'],
        s=2, alpha=0.25,
        color=CLUSTER_COLORS[cl],
        label=CLUSTER_LABELS[cl],
        rasterized=True,
    )

# ── Primer plano: golpes verificados ──
for cl in sorted(df['cluster'].unique()):
    mask = (df['cluster'] == cl) & (df['es_golpe'] == 1)
    sub = df[mask]
    if len(sub) == 0:
        continue
    ax.scatter(
        sub['umap_x'], sub['umap_y'],
        s=250, marker='*',
        color='#e74c3c',
        edgecolors='black', linewidths=0.8,
        zorder=5,
        label=f'Golpe verificado (cl={cl}, n={len(sub)})' if cl == sorted(df.loc[df['es_golpe']==1, 'cluster'].unique())[0] else '_nolegend_',
    )

# Leyenda unificada para golpes
n_total_golpes = int(df['es_golpe'].sum())
from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0],[0], marker='o', color='w', markerfacecolor=CLUSTER_COLORS[k],
           markersize=8, label=CLUSTER_LABELS[k])
    for k in sorted(CLUSTER_COLORS)
]
legend_elements.append(
    Line2D([0],[0], marker='*', color='w', markerfacecolor='#e74c3c',
           markeredgecolor='black', markersize=14,
           label=f'Golpe verificado (n={n_total_golpes})')
)

ax.legend(handles=legend_elements, fontsize=11, loc='best')
ax.set_xlabel('UMAP dim 1', fontsize=12)
ax.set_ylabel('UMAP dim 2', fontsize=12)
ax.set_title(
    f'UMAP de ángulos biomecánicos ({X.shape[0]:,} jugador×frame)\n'
    f'Coloreado por cluster KMeans · Estrellas rojas = golpes verificados manualmente',
    fontsize=13
)
ax.grid(alpha=0.2)
plt.tight_layout()
plt.show()
```

```python
# Distribución de golpes por cluster
dist = df[df['es_golpe']==1]['cluster'].value_counts().sort_index()

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# -- Barras: conteo de golpes por cluster --
bars = axes[0].bar(
    [f'Cluster {int(k)}' for k in dist.index],
    dist.values,
    color=[CLUSTER_COLORS[int(k)] for k in dist.index],
    edgecolor='black', linewidth=0.8, alpha=0.9
)
for bar, v in zip(bars, dist.values):
    axes[0].text(bar.get_x() + bar.get_width()/2, v + 0.3,
                str(v), ha='center', fontsize=12, fontweight='bold')
axes[0].set_title('Golpes verificados por cluster', fontsize=12)
axes[0].set_ylabel('Cantidad')

# -- Scatter: solo golpes, colorados por cluster, más visible --
for cl in sorted(df['cluster'].unique()):
    mask_bg  = (df['cluster'] == cl) & (df['es_golpe'] == 0)
    mask_gol = (df['cluster'] == cl) & (df['es_golpe'] == 1)
    axes[1].scatter(
        df.loc[mask_bg,  'umap_x'], df.loc[mask_bg,  'umap_y'],
        s=2, alpha=0.12, color=CLUSTER_COLORS[cl], rasterized=True
    )
    sub = df[mask_gol]
    if len(sub):
        axes[1].scatter(
            sub['umap_x'], sub['umap_y'],
            s=320, marker='*',
            color=CLUSTER_COLORS[cl],
            edgecolors='black', linewidths=1.2,
            zorder=5,
            label=f'Cluster {cl}  (n={len(sub)})'
        )
axes[1].legend(fontsize=10)
axes[1].set_xlabel('UMAP dim 1', fontsize=11)
axes[1].set_ylabel('UMAP dim 2', fontsize=11)
axes[1].set_title('Golpes por cluster en el espacio UMAP\n(estrella = color del cluster al que pertenece)', fontsize=11)
axes[1].grid(alpha=0.2)

plt.tight_layout()
plt.show()

print("Distribución de golpes verificados por cluster:")
total_cluster = df['cluster'].value_counts().sort_index()
for cl in sorted(df['cluster'].unique()):
    n_g = int((df['es_golpe']==1).sum() and dist.get(cl, 0))
    n_g = int(dist.get(cl, 0))
    n_t = int(total_cluster.get(cl, 0))
    pct = n_g / n_t * 100 if n_t > 0 else 0
    print(f"  Cluster {cl}: {n_g} golpes de {n_t:,} totales  ({pct:.2f}%)")
```
